The Airline Passengers dataset describes the total number of airline passengers over a period of time.
The units are a count of the number of airline passengers in thousands. There are 144 monthly observations from 1949 to 1960.
Step 1. Looking at the graph of raw observations
Observation: Reviewing the line plot, it suggests that there may be a linear trend. There is also seasonality, but the amplitude (height) of the cycles appears to be increasing, suggesting that it is multiplicative

```python
from pandas import read_csv
from matplotlib import pyplot
series = read_csv('airline-passengers.csv', header=0, index_col=0)
series.plot()
pyplot.show()
```

Step 2. Assuming a multiplicative model.
The example code below decomposes the airline passengerâ€™s dataset as a multiplicative model.

```python
from pandas import read_csv, to_datetime
from matplotlib import pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Reading the data
series = read_csv('airline-passengers.csv', header=0, index_col=0)

# Ensuring the index is a datetime index
series.index = to_datetime(series.index)

# Checking for and handle any missing data
series = series.dropna()

# Performing seasonal decomposition
result = seasonal_decompose(series, model='multiplicative')

# Plotting the decomposition results
result.plot()
plt.show()
```

We can see that the trend and seasonality information extracted from the series does seem reasonable. The residuals are also interesting, showing periods of high variability in the early and later years of the series.
